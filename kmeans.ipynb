## **Clustering por K-means**

#### **Carga de librer√≠as & datos**

```python
import pandas as pd, matplotlib.pyplot as plt, seaborn as sns, numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('BetterLife_cleaned.csv')
```

#### **K-Means**

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Seleccionar las columnas num√©ricas para el clustering
numeric_columns = df.select_dtypes(include=['float64']).columns
data = df[numeric_columns]

# Escalar los datos
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Aplicar K-means
kmeans = KMeans(n_clusters = 4, random_state = 50)  # Cambia n_clusters seg√∫n el n√∫mero de grupos deseados
df['Cluster'] = kmeans.fit_predict(scaled_data)

df.sort_values(by='Cluster', ascending= True, inplace=True)

# Mostrar los pa√≠ses con sus respectivos clusters
print(df[['Country', 'Cluster']])

fig, ax = plt.subplots(figsize=(40, 6))
sns.scatterplot(data=df, x='Country', y='Cluster', hue='Cluster', palette='Set1', ax=ax)
None
```

#### **Exportaci√≥n del dataset**

```python
df.to_csv('BetterLife_clustered.csv', index = False)
```
